```python
import numpy as np 
import matplotlib.pyplot as plt

def _fitness(x):
    #x = np.array(x)
    if x > - 11 and x < 11:
        y = (x**2+x)*np.cos(2*x) + x**2
        return round(y,6)
    else:
        return 0
fitness = np.vectorize(_fitness)
```
